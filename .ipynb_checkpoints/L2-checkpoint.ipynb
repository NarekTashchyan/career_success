```python
print('Names of the survivors')
for name, survived in zip(df['Name'], df['Survived']):
    if survived == 1:
        print(name)
```

```python
print('Names of the passangers older than 25')
for name, age in zip(df['Name'], df['Age']):
    if age > 25:
        print(name)
```
